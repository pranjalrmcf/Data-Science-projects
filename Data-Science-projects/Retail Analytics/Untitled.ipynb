```python
for column in num.columns:
    plt.figure()
    sns.boxplot(num[column])
```

```python
for column in num.columns:
    q1 = num[column].quantile(0.25)
    q3 = num[column].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    num[column] = np.where(num[column]>upper, np.nan, num[column])
    num[column] = np.where(num[column]<lower, np.nan, num[column])
```
