```python
import numpy as np
import pandas as pd
import tensorflow
import keras
```

```python
data = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
data.head()
```

```python
data['DEATH_EVENT'].value_counts()
```

```python
data.isnull().sum()
```

```python
X = data.drop(['DEATH_EVENT'], axis = 1)
Y = data['DEATH_EVENT']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns = X.columns)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import InputLayer,Dense
input_neurons = X_train.shape[1]
number_of_hidden_layers = 2
neuron_hidden_layer_1 = 10
neuron_hidden_layer_2 = 5
output_neuron = 1
model = Sequential()
model.add(InputLayer(input_shape=(input_neurons)))
model.add(Dense(units = neuron_hidden_layer_1, activation="relu"))
model.add(Dense(units = neuron_hidden_layer_2, activation="relu"))
model.add(Dense(units = output_neuron, activation = "sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss = "binary_crossentropy", optimizer = "Adam", metrics = "accuracy")
```

```python
model_history = model.fit(X_train,Y_train, validation_data=(X_test,Y_test), epochs = 50)
```

```python
prediction = model.predict(X_test)
```

```python
prediction_classes = np.array(prediction)
prediction_classes[prediction_classes>=0.5] = 1
prediction_classes[prediction_classes<0.5] = 0
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(prediction_classes,Y_test)
```
