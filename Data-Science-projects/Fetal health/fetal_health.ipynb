```python
X = df.drop(['fetal_health'], axis = 1)
Y = df[['fetal_health']]
```

```python
for column in num.columns:
    plt.figure()
    sns.boxplot(num[column])
```

```python
for column in num.columns:
    q1 = num[column].quantile(0.25)
    q3 = num[column].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    num[column] = np.where(num[column]<lower, np.nan, num[column])
    num[column] = np.where(num[column]>upper, np.nan, num[column])
```

```python
from sklearn.impute import KNNImputer
knn = KNNImputer()
num = pd.DataFrame(knn.fit_transform(num), columns = num.columns)
```

```python
for column in num.columns:
    plt.figure()
    sns.boxplot(num[column])
```

```python
plt.figure(figsize = (13,13))
sns.heatmap(num.corr())
```

```python
from sklearn.preprocessing import KBinsDiscretizer
bins = KBinsDiscretizer(n_bins=10, strategy='quantile', encode='ordinal')
num_binned = pd.DataFrame(bins.fit_transform(num), columns = num.columns)
```

```python
X_num_binned = pd.concat([num_binned, Y], axis = 1)
```

```python
from sklearn.utils import resample

# Separate Target Classes
df_1 = df[df.fetal_health==1]
df_2 = df[df.fetal_health==2]
df_3 = df[df.fetal_health==3]
 
# Upsample minority class
df_2_upsampled = resample(df_2, 
                                 replace=True,     # sample with replacement
                                 n_samples=1601,    # to match majority class
                                 random_state=123) # reproducible results

df_3_upsampled = resample(df_3, 
                                 replace=True,     # sample with replacement
                                 n_samples=1601,    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_1, df_2_upsampled, df_3_upsampled])
```

```python
df_upsampled.fetal_health.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
Y_pred = log_reg.predict(X_test)
```

```python
print(log_reg.score(X_train,Y_train))
print(log_reg.score(X_test,Y_test))
```
