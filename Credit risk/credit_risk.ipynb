### Dropping features which have more than 20% null values

```python
for col in data.columns:
    if data[col].isnull().mean() > 0.2:
        data.drop([col], axis = 1, inplace = True)
```

```python
data.info()
```

```python
num = data.select_dtypes(include = 'number')
char = data.select_dtypes(include = 'object')
```

### Imputation of NULL values in num dataframe

```python
for col in num.columns:
    plt.figure()
    sns.distplot(num[col])
```

```python
for col in num.columns:
    num[col].fillna(num[col].median(), inplace = True)
```

```python
num.isnull().sum()
```

```python
for col in num.columns:
    plt.figure()
    sns.boxplot(num[col])
```

```python
for col in num.columns:
    q1 = num[col].quantile(0.25)
    q3 = num[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    num[col] = np.where(num[col]<lower, lower, num[col])
    num[col] = np.where(num[col]>upper, upper, num[col])
```

```python
for col in num.columns:
    plt.figure()
    sns.boxplot(num[col])
```
